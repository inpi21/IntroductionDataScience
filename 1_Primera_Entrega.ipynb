####Descripción del problema
Para la comunicación con los clientes, el banco cuenta con canales como la página web del banco, la aplicación del banco, chat por WhatsApp, Mensajes de texto, llamadas telefónicas, atención presencial entre otros, los cuales se están utilizando sin aprovechar conocimientos que ya se tienen como las preferencias de los clientes, efectividad previa por clientes de los canales y canales que frecuenta.

####Solución actual
En la actualidad, la selección del canal para comunicar los mensajes a los clientes, se hace de forma experimental.

Dado que se ha experimentado con varios canales, se tiene información de efectividad y de costos de los canales.

La estretegia actual permite postular otras que permitan mejorar el uso de los recursos de comunicación y que sean de mayor aceptación en el cliente.


###1.2.2 Defina los objetivos de negocio

Desarrollar una estrategía de selección de canal para la comunicación a clientes que permita mejorar el uso de los recursos económicos disponibles y considerar aspectos como:

*   Canales previos utilizados con el cliente y su efectividad,
*   Comportamientos del cliente,
*   Medios y canales que frecuenta con el banco.

Se plantearon para el proyecto dos propósitos de comunicación con clientes a tener en cuenta:

*   Comunicación de oferta crediticia,
*   Promoción de uso de canales digitales.

Se formularon como preguntas de negocio:

¿Cómo podría optimizarse el gasto de comunicación a clientes?

¿Cuál es el mejor canal para comunicar al cliente el mensaje? La optimización debe considerar la efectividad de transmisión de los mensajes y la eficiencia de costos, incluyendo aquellos que se causen por procesos adicionales generados de los mensajes comunicados, por ejemplo una visita adicional para aprobación de créditos.

¿Cuánto podría optimizarse el monto que se invierte hoy en comunicar estos mensajes a los clientes?
